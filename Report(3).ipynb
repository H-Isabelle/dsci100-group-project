```r
library(readr)
driveurl <- "https://drive.google.com/uc?export=download&id=1Mw9vW0hjTJwRWx0bDXiSpYsO3gKogaPz"
players <- read_csv(url(driveurl))
players
```

The data we are analyzing is the "players.csv" file. 

- *experience*: represents experience as self-reported by each individual from the categories of "Beginner", "Amateur", "Regular", "Pro", and "Veteran". The most common ranking appears to be "Amateur". 

```r



experience_stats <- players |>
group_by(experience) |>
summarize(count = n()) |>
mutate(percent = (count / 196) * 100) 
experience_stats 


exp_sum_plot <- ggplot(experience_stats, aes(x = experience, y = percent, fill = experience)) + geom_bar(stat = "identity",position = "dodge") + labs(x = "Reportd Experience Level", y = "Percentage", title = "Frequency of players by experience level")
exp_sum_plot
```

- *subscribe*: a logical variable indicating whether or not this person has subscribed to the email list. The majority of people have subscribed. 

```r
subscription_plot <- ggplot(players, aes(subscribe, fill = subscribe)) + geom_bar(position = "dodge") + labs (x = "Have they subscribed?", y = "Frequency")
subscription_plot

sub_stats <- players |>
group_by(subscribe) |>
summarize(count = n()) |>
mutate(percent = (count/196)*100)

sub_stats
                            
```

- *hashedEmail*: a hashed version of the player's email address. This variable is individualized to each player, so it will not be useful in our analysis.

- *played_hours*: The number of hours an individual has played on the server, rounded to the nearest tenth. One problem with this dataset is that it has an overrepresentation of players who haven't yet played on the server, or who have played a very low amount.

```r
hours_histo <- ggplot(players, aes(played_hours)) + geom_histogram() + labs(x = "Hours played per player", y = "Frequency", title = "Frequency of # of hours played")
#hours_histo

hours_stats <- players |>
group_by(played_hours) |>
summarize(count = n()) |>
mutate(percent = (count/196)*100) 

hours_morethanthree <- hours_stats |>
filter(played_hours >= 3.0) 


#hours_stats
hours_morethanthree
19/196
```

-*name*: A name chosen by the player from a randomly-generated list upon signup, which also act's as a player's username within the server. 

-*gender*: The player's gender. The majority of players are male. 

```r
gender_plot <- ggplot(players, aes(gender, fill = gender)) + geom_bar(position = "dodge") + labs (x = "Player gender", y = "Frequency", title = "Player gender by frequency")
gender_plot

gender_stats <- players |>
group_by(gender) |>
summarize(count = n()) |>
mutate(percent = (count/196)*100)

gender_stats
```

-*age*: The player's age. The majority of players are in their late teens to early twenties. I've filtered out any significant outliars age-wise to make the plot more readable. 

```r
players_filter <- filter(players, age < 51)
age_histo <- ggplot(players_filter, aes(age)) + geom_histogram() + labs(x = "Age", y = "Frequency")
age_histo

age_stats <- players |>
filter(age <= 89) |>
summarize(mean_age = mean(age), median_age = median(age))

age_stats
```
